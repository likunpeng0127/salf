```python
# ========== æ­¥éª¤ 7: æ˜¾ç¤ºç»“æœ ==========
print('ğŸ–¼ï¸  æ˜¾ç¤ºå¤„ç†ç»“æœ...\n')

from PIL import Image
import matplotlib.pyplot as plt
import os

# è·å–æ‰€æœ‰æ–‡ä»¶
input_files = sorted([f for f in os.listdir('inputs') if f.endswith(('.png', '.jpg', '.jpeg'))])
all_output_files = sorted(os.listdir('outputs'))
output_files = sorted([f for f in all_output_files if f.endswith('.png')])

print(f'ğŸ“Š æ‰¾åˆ° {len(input_files)} å¼ åŸå›¾')
print(f'ğŸ“Š outputs æ–‡ä»¶å¤¹ä¸­æœ‰ {len(all_output_files)} ä¸ªæ–‡ä»¶')
print(f'ğŸ“Š å…¶ä¸­ {len(output_files)} ä¸ªæ˜¯ PNG å›¾ç‰‡\n')

# å¦‚æœæ²¡æœ‰è¾“å‡ºæ–‡ä»¶ï¼Œæ˜¾ç¤ºè­¦å‘Š
if len(output_files) == 0:
    print('âŒ é”™è¯¯: outputs æ–‡ä»¶å¤¹ä¸­æ²¡æœ‰æ‰¾åˆ°ç»“æœå›¾ç‰‡ï¼')
    print('å¯èƒ½çš„åŸå› :')
    print('  1. æ­¥éª¤6çš„æ¨ç†è¿‡ç¨‹å‡ºé”™äº†')
    print('  2. è¯·æ£€æŸ¥æ­¥éª¤6æ˜¯å¦æœ‰çº¢è‰²é”™è¯¯ä¿¡æ¯')
    print('  3. ç¡®è®¤å·²ç»æ­£ç¡®å®‰è£…äº† PyTorch 1.7.1')
    print('\nğŸ’¡ å»ºè®®: é‡æ–°è¿è¡Œæ­¥éª¤6ï¼Œä»”ç»†æŸ¥çœ‹è¾“å‡ºä¿¡æ¯')
else:
    print('ğŸ“‹ è¾“å‡ºæ–‡ä»¶:')
    for f in output_files[:5]:  # åªæ˜¾ç¤ºå‰5ä¸ª
        print(f'   {f}')
    if len(output_files) > 5:
        print(f'   ... è¿˜æœ‰ {len(output_files)-5} ä¸ªæ–‡ä»¶')
    print()

    # æ˜¾ç¤ºæ¯ç»„å¯¹æ¯”
    success_count = 0
    for idx, inp_file in enumerate(input_files, 1):
        # å°è¯•å¤šç§æ–‡ä»¶ååŒ¹é…æ–¹å¼
        base_name = os.path.splitext(inp_file)[0]
        possible_names = [
            inp_file.replace('.jpg', '.png').replace('.jpeg', '.png'),  # ç›´æ¥æ›¿æ¢æ‰©å±•å
            base_name + '.png',  # åŸºç¡€å + .png
            base_name + '_vis.png',  # æœ‰äº›ç‰ˆæœ¬ä¼šæ·»åŠ  _vis åç¼€
        ]

        out_file = None
        for possible in possible_names:
            if possible in output_files:
                out_file = possible
                break

        if out_file is None:
            print(f'âš ï¸  æœªæ‰¾åˆ° {inp_file} çš„è¾“å‡ºç»“æœ')
            print(f'    å°è¯•è¿‡çš„æ–‡ä»¶å: {possible_names}')
            continue

        # åˆ›å»ºå­å›¾
        fig, axes = plt.subplots(1, 2, figsize=(14, 7))
        fig.suptitle(f'å›¾ç‰‡ {idx}/{len(input_files)}: {inp_file}', fontsize=14, fontweight='bold')

        # æ˜¾ç¤ºåŸå›¾
        img_input = Image.open(f'inputs/{inp_file}')
        axes[0].imshow(img_input)
        axes[0].set_title('åŸå›¾', fontsize=12)
        axes[0].axis('off')

        # æ˜¾ç¤ºè§£æç»“æœ
        img_output = Image.open(f'outputs/{out_file}')
        axes[1].imshow(img_output)
        axes[1].set_title(f'äººä½“è§£æç»“æœ ({dataset.upper()})', fontsize=12)
        axes[1].axis('off')

        plt.tight_layout()
        plt.show()
        print(f'  âœ… {inp_file} â†’ {out_file}')
        print(f'  å°ºå¯¸: {img_input.size[0]}x{img_input.size[1]} åƒç´ ')
        print()
        success_count += 1

    print(f'âœ… æˆåŠŸæ˜¾ç¤º {success_count}/{len(input_files)} ç»„ç»“æœï¼')
```
